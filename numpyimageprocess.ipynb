```python
# An image can be considered an array of numbers
import numpy as np

from PIL import Image
from IPython.display import display

im=Image.open("girl.jpg")
display(im)

#converting the image to numpy array

array = np.array(im)
array.dtype
array.shape

# 'uint8' mean 2 to the power 8. This 256 (actually 255 since we start from 0). Here 0-black and 255-white

#creating a mask array with same shape

mask = np.full(array.shape,255)
print(mask)

invert=mask-array


print(invert)
invert=invert.astype(np.uint8)

#converting the array to a display on jupyter

display(Image.fromarray(invert))
```
